```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
salary= pd.read_csv("C:/Users/DELL/Downloads/Salary_Data.csv")
```

```python
salary.head()
```

```python
salary.describe()
```

```python
plt.title('Salary Distribution Plot')
sns.distplot(salary['Salary'])
plt.show()
```

```python
plt.scatter(salary['YearsExperience'], salary['Salary'], color = 'lightcoral')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X = salary.iloc[:, :1]  # independent
y = salary.iloc[:, 1:]  # dependent
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
y_pred_train = regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np


# Calculate metrics
mse = mean_squared_error(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_test)

# Print metrics
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared Score (R):", r2)

if r2 < 0.3:
    print("Model Diagnosis: Underfitting  The model is not learning enough from the data.")
elif r2 >= 0.3 and r2 < 0.7:
    print("Model Diagnosis: Bad/Weak  The model captures some patterns but not reliably.")
elif r2 >= 0.7 and r2 < 0.9:
    print("Model Diagnosis: Good  The model fits the data reasonably well.")
elif r2 >= 0.9 and r2 <= 1.0:
    print("Model Diagnosis: Excellent  The model fits the data very well.")
else:
    print("Model Diagnosis: Overfitting or Error  R is unusually high or invalid.")
```
