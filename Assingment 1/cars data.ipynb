```python
from datetime import date

current_year = date.today().year
cars_data['Car_Age'] = current_year - cars_data['Year']
cars_data.head()
```

```python
cars_data['IS_Automatic'] = cars_data['Transmission']
cars_data.head()
```

```python
cars_data['IS_Automatic'] = cars_data['Transmission'].apply(lambda x: 1 if x == 'Manual' else 0)
cars_data.head()
```

```python
cars_data['Trust'] = cars_data['Owner_Type'].apply(lambda x: 1 if x == 'First' else 0)
cars_data.head()
```

```python
cars_data['Brand'] = cars_data.Name.str.split().str.get(0)
```

```python
cars_data['Model'] = cars_data.Name.str.split().str.get(1) + cars_data.Name.str.split().str.get(2)
```

```python
fuel_counts = cars_data['Fuel_Type'].value_counts()

plt.bar(fuel_counts.index, fuel_counts.values, color='blue')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Fuel Type Distribution')
plt.show()
```

```python
model_counts = cars_data['Name'].value_counts()

cars_data['Brand'] = cars_data['Name'].apply(lambda x: x.split()[0])

brand_counts = cars_data['Brand'].value_counts()

plt.pie(brand_counts, labels=brand_counts.index, startangle=140)
plt.title('Car Distribution by Brand')
plt.axis('equal')
plt.show()
```
